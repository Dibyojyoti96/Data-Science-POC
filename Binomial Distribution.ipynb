# Binomial Distribution

## Example:

        A bank issues credit  card to customers under the scheme of Mastercard. Based on the past datd, the bank has found that 60% of all accounts pay on time following the bill. If a sample of 7 accounts is selected at random from the current database, construct the Binomial Probability of accounts paying on time.

```python
# importing libraries
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# probability of individual success
p = 0.6
# total number of trials
n = 7
# creating a sample space for which I want to calculate the probability
s = np.arange(0,8)
```

```python
s
```

```python
# 0 means none pay on time
# 7 means all pay on time
```

```python
# calculating the binomial probability for every possibilities
bino = stats.binom.pmf(s,n,p)
bino
```

## Plotting the graph

```python
plt.plot(s,bino,"o-")
plt.title("Binomial Probability")
plt.xlabel("Number of Successes")
plt.ylabel("Probability of success")
```

so, the number of people expected to pay on time will be computed by:
E(Y)= n.p
Therefore, E(Y) = 7*0.6 = 4.2.
Thus, the number of people expected to pay on time is 4.
