## Applying functions in a dataset

```python
store_sales =pd.read_csv("store_sales.csv")
```

```python
store_sales.head()
```

```python
def bonus_(sale_num):
    return ["Eligible" if x>18 else "Not Eligible" for x in sale_num]
```

```python
# applying to check in which all months,stores were eligile for bonus

store_sales.iloc[:,2:].apply(bonus_).head()
```

```python
#finding out the average sales for each month

store_sales.groupby("city").agg("mean")
```

## Finding out missing values and replacing it

```python
#to find out any category has a missing value

df.isnull().sum()
```

```python
#Purpose category has the most missing values
# need to find out the most common value in the Purpose category
df["PURPOSE*"].value_counts()
```

```python
#most common value is "Meeting"
#replacing the missing values with the most common value

df=df.fillna({"PURPOSE*":"Meeting"})
```

```python
df.isnull().sum()
df["PURPOSE*"].value_counts()
```

## Matrix Operation 

```python
#computing month on month growth in sales

sales_matrix = np.array(store_sales.iloc[:,2:])
sales_matrix
```

```python
sales_matrix_lag =sales_matrix[:,1:]
```

```python
sales_matrix_lag
```

```python
sales_matrix_lag = np.c_[sales_matrix_lag,np.zeros(100)]
sales_matrix_lag
```

```python
sales_diff = sales_matrix_lag - sales_matrix
```

```python
perc_change = pd.DataFrame(sales_diff/sales_matrix)
```

```python
perc_change.head()
```
